```python
Answers = pd.read_csv('Answers.csv',engine='python',delimiter = ',')
```

```python
Answers.rename(columns={'Id':'InitialId'}, inplace=True)
Answers.rename(columns={'ParentId':'Id'}, inplace=True)
Answers.rename(columns={'Body':'Answer'}, inplace=True)
```

```python
Answers
```

```python
tags = pd.read_csv('Tags.csv',engine='python',delimiter = ',')
```

```python
sql_table = tags[tags['Tag'] == 'sql']
```

```python
# sql_table.join(Answers)
inter = pd.merge(sql_table, Answers, on='Id', how='inner')
```

```python
# Questions
pd.read_csv('Questions.csv',delimiter = ',',encoding='latin-1')
```

```python
questions = pd.read_csv('Questions.csv',delimiter = ',',encoding='latin-1')
```

```python
questions
```

```python
result = pd.merge(inter, questions, on='Id', how='inner')
```

```python
result.rename(columns={'Body':'question'}, inplace=True)
```

```python
result = result[['question','Answer','Id']]
result['Source'] = 'Stackoverflow'
result.rename(columns={'Id':'link'}, inplace=True)
result = result[['Source','question','Answer',"link"]]
```

```python
# df['col'] = 'str' + df['col'].astype(str)
result['link'] = "https://stackoverflow.com/questions/" + result["link"].astype(str) 
```
