```python
import pandas as pd
import re
```

```python
pd.read_csv('Answers.csv',engine='python',delimiter = ',')
```

```python
Answers = pd.read_csv('Answers.csv',engine='python',delimiter = ',')
```

```python
Answers.rename(columns={'Id':'InitialId'}, inplace=True)
Answers.rename(columns={'ParentId':'Id'}, inplace=True)
Answers.rename(columns={'Body':'Answer'}, inplace=True)
```

```python
tags = pd.read_csv('Tags.csv',engine='python',delimiter = ',')
```

```python
sql_table = tags[tags['Tag'] == 'sql']
```

```python
sql_table
```

```python
# Questions
pd.read_csv('Questions.csv',delimiter = ',',encoding='latin-1')
```

```python
questions = pd.read_csv('Questions.csv',delimiter = ',',encoding='latin-1')
```

```python
q_and_tags_sql = pd.concat([sql_table, questions], axis=1, join='inner')
q_and_tags_sql.rename(columns={'Body':'question'}, inplace=True)
```

```python
result = pd.concat([q_and_tags_sql, Answers], axis=1, join='inner')
result = result[['question','Answer']]
result['Source'] = 'Stackoverflow'
```

```python
result = result[['Source','question','Answer']]
```

```python
result.to_csv('StackOverflow.csv',index=False)
```
